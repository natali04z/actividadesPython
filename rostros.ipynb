```python
import cv2
```

```python
img_rostro = cv2.imread('imagenes/input_image2.jpg')
```

```python
gray_img =cv2.cvtColor(img_rostro, cv2.COLOR_BGR2GRAY)
```

```python
gray_img.shape
```

```python
cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
if cascade.empty():
    print('El clasificador no se ha cargado correctamente')
```

```python
face = cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=7, minSize=(40,40))
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img_rostro, (x,y), (x+w, y+h), (0, 255, 0), 2)
```

Captura de video en tiempo real

```python
video_capture = cv2.VideoCapture(0)
```

```python
cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
if cascade.empty():
    print('El clasificador no se ha cargado correctamente')
```

```python
def detect_faces(frame):
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = cascade.detectMultiScale(gray_frame, 1.1, 5, minSize=(40, 40))
    for(x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0),4)
        return frame
```

```python
while True:
    ret, frame = video_capture.read()
    if not ret:
        break
    frame= detect_faces(frame)
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    else:
        print('Error al leer la fotografa del video')

        frame.release()
        cv2.destroyAllWindows()
```
